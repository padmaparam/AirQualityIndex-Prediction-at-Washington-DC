```python
from bs4 import BeautifulSoup
import requests
import sys
from bs4 import BeautifulSoup
```

```python
def get_meta_data(month,year):
    htmlfile = open('data/html/{}/{}.html'.format(year,month),'rb')
    plain_text = htmlfile.read()
    
    tempdata = []
    finaldata = []
    
    soup = BeautifulSoup(plain_text, 'lxml')
    for table in soup.findAll('table', {'class': 'medias mensuales numspan'}):
        for tbody in table:
            for tr in tbody:                
                tempdata.append(tr.get_text())
    return(tempdata)
    rows = len(tempdata)/15
```

```python
num_features = 15
for year in range(2009,2020):
    for  month in range(1,13):        
        data = get_meta_data(month, year)
        rows = round(len(data)/num_features)
        df = pd.DataFrame(columns=data[0:num_features-1])
        for i in range(1,rows-1):
            row_start = i*num_features
            row_end = row_start+num_features-1            
            df.loc[i] = data[row_start:row_end]
            file_name = 'data\html\AQI_{}_{}.csv'.format(month,year)    
            df.to_csv(file_name)
```

```python
combine_data = pd.DataFrame()
for year in range(2009,2020):
    climate_data_year = pd.DataFrame()
    for month in range(1,13):
        file_name_html = 'data\html\AQI_{}_{}.csv'.format(month, year)
        climate_data_month = pd.read_csv(file_name_html)    
        climate_data_month['Date'] = str(month).zfill(2)+ '/'+climate_data_month['Day'].astype(str).str.zfill(2) + '/'+ str(year)
        
        climate_data_year = climate_data_year.append(climate_data_month)  

    file_name_aqi = 'data/aqi/ad_viz_plotaqi_data_DC_{}.csv'.format( year)
    aqi_data_year = pd.read_csv(file_name_aqi)
    combine_data_year = pd.merge(climate_data_year,
                             aqi_data_year, how = 'inner', on = 'Date')
    combine_data = combine_data.append(combine_data_year)
```
